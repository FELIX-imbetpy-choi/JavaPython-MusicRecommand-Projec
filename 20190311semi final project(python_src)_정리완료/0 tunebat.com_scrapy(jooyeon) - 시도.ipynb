```python
from selenium import webdriver
import urllib.request as req
from bs4 import BeautifulSoup
import pandas as pd
```

```python
driver = webdriver.Chrome('./chromedriver')
driver.get('https://tunebat.com/Search?q=label:YG-Entertainment')
```

```python
html = driver.page_source
soup = BeautifulSoup(html,'html.parser')

name = soup.select('#react_0HLKRH49Q9RLS > div > div > div.searchResultList.row.main-row > div > div > a > div > div > div.row.search-artist-name')
title = soup.select('#react_0HLKRH49Q9RLS > div > div > div.searchResultList.row.main-row > div > div > a > div > div > div.row.search-track-name')
key = soup.select('#react_0HLKRH49Q9RLS > div > div > div.searchResultList.row.main-row > div > div > a > div > div > div > div > div.row.search-attribute-value')

nameLists = [] 
titleLists = [] 
keyLists = [] 
bpmLists=[]

for namelst in name:
    nameLists.append(namelst.text)
    
for titlelst in title:
    titleLists.append(titlelst.text)
    
for i in range(0,141):
    if i % 3 == 0:
        keyLists.append(soup.select('#react_0HLKRH49Q9RLS > div > div > div.searchResultList.row.main-row > div > div > a > div > div > div > div > div.row.search-attribute-value')[i].text)
    if i % 3 == 2:
        bpmLists.append(soup.select('#react_0HLKRH49Q9RLS > div > div > div.searchResultList.row.main-row > div > div > a > div > div > div > div > div.row.search-attribute-value')[i].text)
    
print(nameLists)
print(titleLists)
print(keyLists)
print(bpmLists)
```
