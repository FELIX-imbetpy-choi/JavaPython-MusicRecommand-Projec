```python
driver = webdriver.Chrome('./chromedriver.exe')
driver.get('https://tunebat.com/Search?q=label:%EC%98%A4%EA%B0%90%EC%97%94%ED%84%B0%ED%85%8C%EC%9D%B8%EB%A8%BC%ED%8A%B8') # 새창이 하나 뜰 것
```

```python
lists = []
a = soup.find_all('a')
for i in a:
    b = i.attrs['href']
    if b.split('/')[1] == 'Info':
        lists.append(b)
print(lists)
```

```python
# 해당 페이지에서 노래 하이퍼링크 있는 구역만 가져오기
lists = []
a = soup.find_all('a')
for i in a:
    b = i.attrs['href']
    if b.split('/')[1] == 'Info':
        lists.append(b)

        
urlbase = 'https://tunebat.com'
info1 = []
info2 = []
counts = 0
for i in lists:
    workurl = urlbase + i
    
    # 해당 URL로 연결
    driver.get(workurl)
    time.sleep(3)
    
     # 가수, 제목, 키내용, 키, Camelot내용, Camelot, 재생시간, Duration, BPM숫자 ,BPM'
    info1.append(driver.find_element_by_xpath('//*[@id="infoPageMain"]/div/div[1]').text)
    # Energy Danceability Happiness Loudness\n38 32 29 -7 dB\nAcousticness Instrumentalness Liveness Speechiness\n60 0 12 3
    info2.append(driver.find_element_by_xpath('//*[@id="infoPageMain"]/div/div[3]').text)
    counts += 1
    print('\n----------------------\n',counts,'\n----------------------\n',info1[counts-1],'\n',info2[counts-1])
    
    if counts >= len(lists):
        print('1페이지 끝')
        driver.get('https://tunebat.com/Search?q=label:%EC%98%A4%EA%B0%90%EC%97%94%ED%84%B0%ED%85%8C%EC%9D%B8%EB%A8%BC%ED%8A%B8&page=2')
    else:
        driver.get('https://tunebat.com/Search?q=label:%EC%98%A4%EA%B0%90%EC%97%94%ED%84%B0%ED%85%8C%EC%9D%B8%EB%A8%BC%ED%8A%B8')

```

```python
driver = webdriver.Chrome('./chromedriver.exe')
driver.get('https://tunebat.com/Search?q=label:SM-Entertainment&page=1') # 새창이 하나 뜰 것
```

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
```

```python
lists = []
a = soup.find_all('a')
for i in a:
    b = i.attrs['href']
    if b.split('/')[1] == 'Info':
        lists.append(b)
print(lists)            
```

```python
# SM 엔터테이먼트 검색 시 모든 결과 가져오기
# 해당 페이지에서 노래 하이퍼링크 있는 구역만 가져오기
info1 = [] # 정보 1
info2 = [] # 정보 2
number = 0 # 전체 데이터 확인용

# SM은 총 44페이지까지 있는 것을 확인 후 범위 설정
for j in range(1,45):
    # 페이지 이동 할 때 넘어간 페이지 파싱(parser)
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    lists = []
    a = soup.find_all('a') # 중간에 광고 때문에 실질적으로 음악 관련 하이퍼링크가 있는 셀 주소 가져오기
    for i in a:
        b = i.attrs['href'] # a태그 href 빼오기
        if b.split('/')[1] == 'Info':
            lists.append(b)
            
    counts = 0 # 한 페이지에 있는 셀 수 카운트
    urlbase = 'https://tunebat.com'
    for i in lists:
        workurl = urlbase + i

        # 해당 URL로 연결
        driver.get(workurl)
        time.sleep(3)
               
         # 가수, 제목, 키내용, 키, Camelot내용, Camelot, 재생시간, Duration, BPM숫자 ,BPM' 이런 순서
        info1.append(driver.find_element_by_xpath('//*[@id="infoPageMain"]/div/div[1]').text)

        # Energy Danceability Happiness Loudness\n38 32 29 -7 dB\nAcousticness Instrumentalness Liveness Speechiness\n60 0 12 3 이런 순서
        info2.append(driver.find_element_by_xpath('//*[@id="infoPageMain"]/div/div[3]').text)
        
        # 확인용
        number += 1
        counts += 1
        print('\n----------------------\n',number,'\n----------------------\n',info1[number-1],'\n',info2[number-1])
        
        # 한 페이지의 마지막 번째까지 돌면 '끝'출력 + 다음페이지로 이동
        if counts == len(lists):
            print('**********%d 페이지 끝**********'%j)
            driver.get('https://tunebat.com/Search?q=label:SM-Entertainment&page={0}'.format(j+1))
        else:
            driver.get('https://tunebat.com/Search?q=label:SM-Entertainment&page={0}'.format(j)) # 다 안돌았으면 전 페이지로 돌아가기(뒤로가기)

```

```python
info2[8].split()
```

```python
# info1
singer = []
title = []
key = []
Camelot = []
Duration = []
BPM = []

# info2

Energy = []
Danceability = []
Happiness = []
Loudness = []
Acousticness = []
Instrumentalness = []
Liveness = []


for i in range(len(info1)):
    temp = info1[i].split('\n')
    
    # 가수
    singer.append(temp[0])
    
    # 제목
    title.append(temp[1])
    
    # 키
    key.append(temp[2])
    
    # 코드
    Camelot.append(temp[4])
    
    # 재생시간
    Duration.append(temp[6])
    
    # BPM
    BPM.append(temp[8])
    
for j in range(len(info2)):
    temp2 = info2[j].split()
    
    # Energy
    Energy.append(temp2[7])
    
    # Danceability
    Danceability.append(temp2[8])
    
    # Happiness
    Happiness.append(temp2[9])
    
    # Loudness
    Loudness.append(temp2[10]+temp2[11])
    
    # Acousticness
    Acousticness.append(temp2[12])
    
    # Instrumentalness
    Instrumentalness.append(temp2[13])
    
    # Liveness
    Liveness.append(temp2[14])
    
```

```python
smenter= pd.DataFrame({
    'singer':singer, 'title':title, 'key':key, 'Camelot':Camelot, 'Duration':Duration, 'BPM':BPM, 
    'Energy':Energy, 'Danceability':Danceability, 'Happiness':Happiness, 'Loudness':Loudness, 'Acousticness':Acousticness,
    'Instrumentalness':Instrumentalness, 'Liveness':Liveness
})

smenter.head()
```

```python
smenter.to_csv('./SM_Entertainment.csv',sep=',',index=False, encoding='utf-8')
```
