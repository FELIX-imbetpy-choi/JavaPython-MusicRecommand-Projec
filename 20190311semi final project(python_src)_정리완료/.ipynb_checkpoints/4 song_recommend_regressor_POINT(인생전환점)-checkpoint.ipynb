# 사용자 곡추천 관련 머신러닝 전환점

사용자 곡추천 관련 머신러닝 첫 논의에서 오디오분석 피쳐를 정답으로 주는 머신러닝을 협의.   
이 결과치가 소비자 취향이 전혀 반영되지 않는다는 것을 알아챔.   
그 후 사용자 플레이 수를 정답으로 주는 현재 파일을 제안.   
이 머신러닝을 추후 서비스에 반영함.    


```python
# Module
import pandas as pd
import numpy as np
from scipy import stats # 확률분포 그래프
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics

import warnings
warnings.filterwarnings('ignore')
```

```python
song = pd.read_csv('./song_data_cluster11.csv')
```

```python
lists = list(song['song_name'])
```

```python
playLists = []
playLists2 = []
for i in range(0,2):
    for j in list(np.random.randint(10, 1499, size=100)):
        a = lists[j]
        playLists.append(a)
    playLists2.append(playLists)
```

```python
df = pd.DataFrame(playLists)
df['song_name2'] = playLists
df = df.rename(columns = {0:'song_name'})
```

```python
df_list = []
for i in range(0,len(playLists)):
    z = song.loc[song['song_name']== playLists[i]]
    df_list.append(z)
```

```python
df_list[0]
```

```python
df_list = []
for i in range(len(playLists)):
    df_list.append(song.loc[song['song_name']== playLists[i]])
```

```python
df2 = pd.DataFrame(columns=list(song.columns))
for i in range(len(df_list)):
    df2 = pd.concat([df2,df_list[i]])

df2 = df2.drop_duplicates(keep='first')
df2
```

```python
df2.drop_duplicates(subset=['song_name'], inplace=True)
```

```python
df2['song_popularity'] = df2['song_popularity'].astype(int)
df2['song_duration_ms'] = df2['song_duration_ms'].astype(int)
df2['key'] = df2['key'].astype(int)
df2['audio_mode'] = df2['audio_mode'].astype(int)
df2['time_signature'] = df2['time_signature'].astype(int)
df2['cluster'] = df2['cluster'].astype(int)
```

```python
df2['point']=np.random.randint(1,100,len(df2))
df2.head()
```

```python
target_col = ['song_popularity','song_duration_ms','acousticness','danceability','energy',
              'instrumentalness','key','liveness','loudness','audio_mode','speechiness','tempo',
              'time_signature','audio_valence','cluster']
df2[target_col].head()
```

```python
train, test = train_test_split(df2,test_size=0.3,random_state=0)
```

```python
# Train data 구성
# 모델 제작용 컬럼과 정답 컬럼을 구분
train_X = train[target_col]
train_Y = train['point']
train_Y = train_Y.astype(float)
train_X.head()
```

```python
train_Y.head()
```

```python
X_test = test[target_col]
Y_test = test['point']
Y_test = Y_test.astype(float)
X_test.head()
```

```python
Y_test.head()
```

```python
feature_one = train_X.values
target = train_Y.values
```

```python
# model 생성
model = RandomForestRegressor(n_estimators=100, n_jobs=-1,random_state=0)
model
```

```python
# Training
model.fit(feature_one, target)
```

```python
# 예측
pre = model.predict(X_test)
print(pre)
```

```python
X_test['point']=pre
X_test.sort_values(by='point',ascending=False)
```
