```python
import pymysql
```

```python
# mysql Connection
conn = pymysql.connect(host = '192.168.101.106',user = 'root',password = '1111',
db = 'honjoc',charset = 'utf8')
```

```python
#Connection으로 부터 cursor생성 - 테이터 베이스 빈줄에 커서가 깜빡 거리고 있는 것이다.
curs = conn.cursor()
```

```python
# SQL 문장
sql = 'select * from music'
curs.execute(sql)
```

```python
# Data Fetch
rows = curs.fetchall()
print(rows)
```

```python
# 커넥션 끊어주기 - 끊지 않으면 5명이상 들어가지 못하기 때문에 항상 끊어줘야함.
conn.close()
```

```python
rowList = list(rows)
```

```python
rowList
```

```python
songlist = []
songlist2 = []

for i in rowList:
    songlist2.append(i)
    rowList = list(songlist2)
    songlist.append(rowList)
print(songlist)
```
